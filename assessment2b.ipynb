###### rising star of the week (top 3 candidate who performed well in that particular week)

```python
df = df.sort_values(by='Rising', ascending=False)
top3 = df.head(3)
print(top3)
```

###### Shining stars of the week (top 3 candidates who has highest geekions)

```python
top3 = df.nlargest(3, 'Previous Geekions')
print(top3)
```

###### Department wise codekata performence (pie chart)

```python
department_scores = df.groupby('Department')['CodeKata Score'].mean()

plt.pie(department_scores, labels=department_scores.index, autopct='%1.1f%%')
plt.title('Department-wise CodeKata Performance')
plt.show()
```

###### Department wise toppers (horizantal bar graph or any visual representations of your choice)

```python
department_scores = df.groupby('Department')['CodeKata Score'].max()

plt.barh(department_scores.index, department_scores, height=0.6)
plt.title('Department-wise Toppers')
plt.xlabel('CodeKata Score')
plt.ylabel('Department')
plt.show()
```
